# Optimal Solar Panel Positioning

The geometric efficiency of a solar panel is a factor in it's energy conversion rate. When the surface normal vector of a flat solar panel is pointing directly at the sun the panel has 100% geometric efficiency. Any deviation from parallel leads to a reduction in collection area and by defenition geometric efficiency.

This notebook is used to explore static solar panel mounting on a south facing wall. My house has East/West facing roof surfaces, and i don't wish to mount my seven 250W panels on the roof. Ideally I'd like to mount them vertically on the south facing wall as this require the simplest mounting solution. This notebook investigates the effect of tilting the panel up above the horizon, in order to find an optimal angle, and compute the efficiency loss in basic vertical mounting. It is then up to me to decide whether to take the loss or design and build more complex mounts.

I use the astral module to get data on the suns position. Numpy and Math modules are used to compute vector projections, averages and matplotlib is used to plot.

```python
from astral import *
import datetime
from pytz import timezone
import pytz
import math
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

city_name = 'Leeds'
a = Astral()
a.solar_depression = 'civil'
city = a[city_name]
UTC = pytz.UTC
elevation=city.solar_elevation() #Current solar elevation
print elevation
noon=city.solar_noon()
print a.solar_azimuth(noon,city.latitude,city.longitude) #South
```

```python
import math
panelNormal=[math.radians(0),math.radians(180)] #elevation, azimuth; vector normal to solar panel plane. 
#([0,180] = mounted vertically on south facing house wall)
print city
currentSolarEl=city.solar_elevation()
currentSolarAz=city.solar_azimuth()
#solarVector=[math.radians(solarEl),math.radians(solarAz)]

def geometricEfficiency(panelNormal,solarVector):
    """Geometric efficiency a vector projection of the suns location onto the
    vector orthogonal to the solar panel surface.
    """
    efficiency=math.cos(panelNormal[1]-solarVector[1])*math.cos(panelNormal[0]-solarVector[0])
    return efficiency
```

```python
#Take average  which is proportional to energy
# collection on that day, 1 being maximum efficiency.
#only loop within solarAzimuth+-89degrees, to avoid negativnumbers. Only average angles above horizon/tree line.


def dailyEfficiency(day,efficiency,panelNormal):
    """Returns solar collection geometric efficiency timeseries, with 6 minute frequency,
     over a 24 hour period. For a given day and panel orientation.
    """
    efficiency=np.zeros(24*10)
    for minute in range(len(efficiency)):
        time=day+datetime.timedelta(minutes=6*minute)
        solarEl=a.solar_elevation(time, city.latitude,city.longitude)
        solarAz=a.solar_azimuth(time, city.latitude,city.longitude)
        solarVector=[math.radians(solarEl),math.radians(solarAz)]
        #print solarVector
        if solarEl>10:
            if solarAz>100:
                if solarAz<260:
                    efficiency[minute]=geometricEfficiency(panelNormal,solarVector)
        else:
            efficiency[minute]=0
    return efficiency

def yearlyEff(yearEff,panelNormal):
    """For a given year (fixed to 2014) and panel orientation, returns a timeseries containing
    the variation in average daily geometric efficiency over a year.
    """
    for day in xrange(len(yearEff)):
        start=datetime.datetime(2014, 1, 1, tzinfo=pytz.UTC)
        date=start+datetime.timedelta(day)+datetime.timedelta(seconds=0)
        dayEff=dailyEfficiency(date,efficiency,panelNormal)
        yearEff[day]=sum(dayEff)/len(dayEff)
    return yearEff

day=datetime.datetime(2014,7,14, tzinfo=pytz.UTC)
efficiency=np.zeros(24*10)
efficiency=dailyEfficiency(day,efficiency,panelNormal)

x=np.arange(0,24,0.1)
plt.plot(x,efficiency)
plt.title("Efficiency over a single summer day")
plt.xlabel("Time (hours)")
plt.ylabel("Geometric efficiency")
    
#loop over 24 hour period; IF solarElevation is <10degrees or solarAzimuth is <100 or >260 then efficiency=0.
```

Now we have all the funtions needed to calculate the optimum panel orientation.

Here is a plot of the geometric efficiency variation curves throughout 2014, for various panel elevations.
You can see that in the summer, the higher the angle the better (up to 60degrees). 
In the winter a lower angle captures more.

```python
yearEff=np.zeros(365)
angles=[0,10,20,30,40,50,60]
for angle in angles:
    #print "Panel angle %d" % angle
    panelNormal=[math.radians(angle),math.radians(180)]
    plt.plot(yearlyEff(yearEff,panelNormal),label=angle)
plt.title("Efficiency variation over a year")
plt.xlabel("days")
plt.ylabel("Efficiency")
plt.legend()
plt.show()
```

Next we look at how the average efficiency over a year varies with panel angle.

```python
yearEff=np.zeros(365)
angles=[0,5,10,15,20,25,30,35,40,45,50,55,60]
angularEff=np.zeros(13)
i=0
for angle in angles:
    yearEff=np.zeros(365)
    #print "Panel angle %d" % angle
    panelNormal=[math.radians(angle),math.radians(180)]
    yearEff=yearlyEff(yearEff,panelNormal)
    angularEff[i]=sum(yearEff)/len(yearEff)
    i+=1
plt.title("Average yearly efficiency with panel elevation")
plt.xlabel("Panel elevation above horizon (degrees)")
plt.ylabel("Average yearly collection efficiency")
plt.plot(angles, angularEff)
plt.show()
```

The most efficient angle to mount on a south facing wall in Leeds,UK is the maxima of this curve; approximately 30 degrees. Giving 22% geometric efficiency, this is an average over a year, 24/7. Perhaps it would be useful to also average over only daylight hours.  

Mounting with a 30degree tilt is 18% more efficient than mounting vertically (0.222/0.188=1.18).
